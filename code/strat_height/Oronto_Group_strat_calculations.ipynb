# Notebook for unifying paleomagnetic sample stratigraphic data from the Nonesuch Formation and Freda Formation 

- paleomagnetic data from Henry, 1977 includes site level data with stratigraphic heights hanging from the Nonesuch-Freda contact.
- paleomagnetic data from Slotznick et al., 2023 includes site/specimen level data with stratigraphic heights for the Potato River Falls section with a different datum in the Copper Harbor Conglomerate. Let's recalculate the heights from the Nonesuch-Freda contact, which is 160 m into the stratigraphy.

- Eventually, we need to  put these stratigraphic heights into the stratigraphic age model of Fuentes et al., 2025 so that we can resample the ages assocaited with the paleomagentic site/specimen data. The contact between the Freda and Nonesuch is at 2050 m in the Fuentes et al., 2025 stratigraphy.


- Henry, S. G., Mauk, F. J., & der Voo, R. V. (1977). Paleomagnetism of the upper keweenawan sediments: The nonesuch shale and freda sandstone. Canadian Journal of Earth Sciences, 14(5), 1128–1138. https://doi.org/10.1139/e77-103

- Slotznick, S. P., Swanson-Hysell, N. L., Zhang, Y., Clayton, K. E., Wellman, C. H., Tosca, N. J., & Strother, P. K. (2023). Reconstructing the paleoenvironment of an oxygenated Mesoproterozoic shoreline and its record of life. GSA Bulletin. https://doi.org/10.1130/B36634.1

- Fuentes, A. J., Fairchild, L. M., Hodgin, E. B., Alemu, T., & Swanson-Hysell, N. (2025). Termination of Laurentia’s rapid plate motion at the start of the Grenvillian orogeny. Journal of Geophysical Research: Solid Earth, 130(8), e2025JB031794. https://doi.org/10.1029/2025JB031794


```python
import pandas as pd
```

```python
def save_data(file_name, site_data):
    # Write the extra row to the file first
    with open(file_name, 'w') as f:
        f.write('tab\tsite\n')  # Writing the extra row

    # Append the DataFrame to the file with the original headers
    site_data.to_csv(file_name, mode='a', index=False, sep='\t')
```

```python
Fuentes2025_Nonesuch_Freda_contact = 2050  # in meters

Slotznick2023_Nonesuch_Freda_contact = 160  # in meters
```

## Henry 1977 Data 

```python
Henry1977_site = pd.read_csv('../../data/pmag_compiled/Henry1977/sites.txt', sep='\t', header=1)
Henry1977_sample = pd.read_csv('../../data/pmag_compiled/Henry1977/samples.txt', sep='\t', header=1)
```

```python
Henry1977_site_heights = (Henry1977_sample
    .groupby('site')['height']
    .mean())

# map the heights to the site dataframe
Henry1977_site = Henry1977_site.merge(Henry1977_site_heights, left_on='site', right_index=True)

# now hang the height from the Fuentes et al., 2025 contact height of 2050 m
Henry1977_site['height'] = Fuentes2025_Nonesuch_Freda_contact + Henry1977_site['height']
```

```python
# write out the updated site file with new heights

save_data('../../data/pmag_compiled/Henry1977/sites_with_height.txt', Henry1977_site)
```

## Slotznick 2023 Data

```python
Slotznick2023_specimen = pd.read_csv('../../data/pmag_compiled/Slotznick2023/specimens.txt', sep='\t', header=1)
Slotznick2023_sample = pd.read_csv('../../data/pmag_compiled/Slotznick2023/samples.txt', sep='\t', header=1)

# map sample height from the sample table to the specimen table
Slotznick2023_specimen = Slotznick2023_specimen.merge(Slotznick2023_sample[['sample', 'height']], on='sample')

# recalculate height relative to Fuentes et al., 2025 contact height
Slotznick2023_specimen['height'] = Slotznick2023_specimen['height'] - Slotznick2023_Nonesuch_Freda_contact + Fuentes2025_Nonesuch_Freda_contact

Slotznick2023_specimen
```

```python
save_data('../../data/pmag_compiled/Slotznick2023/specimens_with_height.txt', Slotznick2023_specimen)
```
